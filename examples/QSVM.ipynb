# Quantum Support Vector Machine on Quantum Annealers (QSVM)

The QSVM makes labels data points as 'infected' or 'uninfected' based on the symptomes and patient's age.

```python
import sys
sys.path.append("../src/")

from QSVM import *
```

## Train QSVM

```python
N = 70
validation_pts = 10

# file path to data
file_path = '../data/Covid_Sym2.csv'

QSVM_object = QSVM(file_path=file_path, train_record_number=N, validation_pts=validation_pts, solver_type="simulated_annealing", verbose=False)
alpha, b = QSVM_object.train_SVM()

# calculate model's metrics: f1_score, accuracy, precision, recall
QSVM_object.calculateKPI()
```

## Make Prediction

```python
age = 20
fever = True
throat_pain = False
taste_smell = False

x_test = [age, fever, throat_pain, taste_smell]
c = QSVM_object.predict(x_test)
prediction = "Negative" if c < 0.50 else "Positive"
```

```python
QSVM_object.printconfusionmatrix()
QSVM_object.plotQSVM()
```
